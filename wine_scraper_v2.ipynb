```python
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException, ElementNotVisibleException
import csv
import random
import time

chop = webdriver.ChromeOptions()
# chop.add_argument("--headless")
driver = webdriver.Chrome(chrome_options=chop)

headers = ['#', 'Wine ID', 'Wine Name', 'Wine Type', 'Wine Origin', 'Winery Name', 'Winery Description', 'Alcohol Level', 'Original Price',
           'Discounted Price', 'Saving', 'Sold Out', 'Estimated Price', 'Collectible', 'Red Wine', 'Green Wine', 'White Wine', 'Giftable',
           'Region', 'Region Info', 'Longitude', 'Latitude', 'Ratings',
           'No. of reviews', 'Winemaker Notes', "WS Review", "WR Review", "RP Review", "CG Review", "W&S Review",
           "WW Review", "WE Review", "TP Review", "JS Review", "D Review", 'JH Review', 'V Review',
           "WS Rating", "WR Rating", "RP Rating", "CG Rating", "W&S Rating",
           "WW Rating", "WE Rating", "TP Rating", "JS Rating", "D Rating", 'JH Rating', 'V Rating']

for j in range(2016, 1960, -1):
    vintage_year_headers = [str(j) + "_PROD_ID", str(j) + "_WS", str(j) + "_WR", str(j) + "_RP", str(j) + "_CG",
                            str(j) + "_W&S", str(j) + "_WW", str(j) + "_WE", str(j) + "_TP", str(j) + "_JS",
                            str(j) + "_D", str(j) + "_JH", str(j) + "_V"]
    headers.extend(vintage_year_headers)

with open('data.csv', 'w', newline='', encoding='utf-8') as f:
    writer = csv.writer(f)
    writer.writerow(headers)
count = 0
for i in range(1, 400000):
    url = "https://www.wine.com/product/" + str(i)
    driver.get(url)
    try:
        driver.find_element_by_css_selector('.errorPage')
        continue
    except NoSuchElementException:
        pass
    count += 1
    wine_name = driver.find_element_by_css_selector("h1.pipName").text
    try:
        discounted_price = driver.find_element_by_css_selector("span.productPrice_price-regWhole").text + "." \
                           + driver.find_element_by_css_selector(".productPrice_price-regFractional").text
    except NoSuchElementException:
        discounted_price = ''
    original_price = driver.find_element_by_css_selector("span.productPrice_price-saleWhole").text + "." \
                       + driver.find_element_by_css_selector(".productPrice_price-saleFractional").text
    try:
        saving = driver.find_element_by_css_selector("span.productPrice_savings-amount").text
    except NoSuchElementException:
        saving = ''

    try:
        sold_out = driver.find_element_by_css_selector(".prodItemStock_soldOut")
        if not sold_out.is_displayed():
            raise NoSuchElementException
        sold_out = True
        original_price = ''
        saving = ''
        discounted_price = ''
        est_price = driver.find_element_by_css_selector(".prodItemStock_soldOut-smallText").text
    except NoSuchElementException:
        sold_out = False

    try:
        wine_varietal = driver.find_element_by_css_selector(".prodItemInfo_varietal").text
    except NoSuchElementException:
        wine_varietal = ''

    try:
        wine_origin = driver.find_element_by_css_selector(".prodItemInfo_originText").text
    except NoSuchElementException:
        wine_origin = ''
    try:
        collectible = driver.find_element_by_css_selector("li.prodAttr_icon-collectible")
        collectible = True
    except NoSuchElementException:
        collectible = False

    try:
        red_wine = driver.find_element_by_css_selector("li.prodAttr_icon-redWine")
        red_wine = True
    except NoSuchElementException:
        red_wine = False

    try:
        giftable = driver.find_element_by_css_selector("li.prodAttr_icon-giftable")
        great_gift = True
    except NoSuchElementException:
        great_gift = False

    try:
        green_wine = driver.find_element_by_css_selector("li.prodAttr_icon-greenWine")
        green_wine = True
    except NoSuchElementException:
        green_wine = False

    try:
        white_wine = driver.find_element_by_css_selector("li.prodAttr_icon-whiteWine")
        white_wine = True
    except NoSuchElementException:
        white_wine = False

    try:
        location = driver.find_element_by_css_selector(".pipWineryImage_map a").get_attribute('href')
        location = location.split("/")[-1].split(',')
        longitude = location[0]
        latitude = location[1]
    except NoSuchElementException:
        longitude = ''
        latitude = ''

    try:
        alcohol_level = driver.find_element_by_class_name('prodAlcoholPercent_inner').text
    except NoSuchElementException:
        alcohol_level = ''

    try:
        rating = driver.find_element_by_css_selector(".averageRating_average").text
    except (NoSuchElementException, ElementNotVisibleException):
        rating = ''
    try:
        no_of_reviews = driver.find_element_by_css_selector(".averageRating_number").text
    except (NoSuchElementException, ElementNotVisibleException):
        no_of_reviews = ''
    try:
        winery_name = driver.find_element_by_css_selector(".pipWinery_headline").text
    except (NoSuchElementException, ElementNotVisibleException):
        winery_name = ''
    try:
        winery_description = driver.find_element_by_css_selector(".pipWinery_content").text
    except (NoSuchElementException, ElementNotVisibleException):
        winery_description = ''
    try:
        winemaker_notes = driver.find_element_by_css_selector(".pipWineNotes div.pipWineNotes_copy").text
    except (NoSuchElementException, ElementNotVisibleException):
        winemaker_notes = ''
    try:
        region_name = driver.find_element_by_css_selector(".productPageContentHead_title").text
    except NoSuchElementException:
        region_name = ''
    try:
        region_info = driver.find_element_by_css_selector(".productPageContent p").text
    except NoSuchElementException:
        region_info = ''

    try:
        ww_review = driver.find_element_by_xpath("//div[@class='pipProfessionalReviews_rating']/span[@class='wineRatings_initials'][text() = 'WW']/parent::div/following-sibling::p").text
    except NoSuchElementException:
        ww_review = ''

    try:
        ws_review = driver.find_element_by_xpath("//div[@class='pipProfessionalReviews_rating']/span[@class='wineRatings_initials'][text() = 'WS']/parent::div/following-sibling::p").text
    except NoSuchElementException:
        ws_review = ''

    try:
        we_review = driver.find_element_by_xpath("//div[@class='pipProfessionalReviews_rating']/span[@class='wineRatings_initials'][text() = 'WE']/parent::div/following-sibling::p").text
    except NoSuchElementException:
        we_review = ''

    try:
        tp_review = driver.find_element_by_xpath("//div[@class='pipProfessionalReviews_rating']/span[@class='wineRatings_initials'][text() = 'TP']/parent::div/following-sibling::p").text
    except NoSuchElementException:
        tp_review = ''

    try:
        cg_review = driver.find_element_by_xpath("//div[@class='pipProfessionalReviews_rating']/span[@class='wineRatings_initials'][text() = 'CG']/parent::div/following-sibling::p").text
    except NoSuchElementException:
        cg_review = ''

    try:
        rp_review = driver.find_element_by_xpath("//div[@class='pipProfessionalReviews_rating']/span[@class='wineRatings_initials'][text() = 'RP']/parent::div/following-sibling::p").text
    except NoSuchElementException:
        rp_review = ''

    try:
        wr_review = driver.find_element_by_xpath("//div[@class='pipProfessionalReviews_rating']/span[@class='wineRatings_initials'][text() = 'WP']/parent::div/following-sibling::p").text
    except NoSuchElementException:
        wr_review = ''

    try:
        d_review = driver.find_element_by_xpath("//div[@class='pipProfessionalReviews_rating']/span[@class='wineRatings_initials'][text() = 'D']/parent::div/following-sibling::p").text
    except NoSuchElementException:
        d_review = ''

    try:
        js_review = driver.find_element_by_xpath("//div[@class='pipProfessionalReviews_rating']/span[@class='wineRatings_initials'][text() = 'JS']/parent::div/following-sibling::p").text
    except NoSuchElementException:
        js_review = ''

    try:
        jh_review = driver.find_element_by_xpath("//div[@class='pipProfessionalReviews_rating']/span[@class='wineRatings_initials'][text() = 'JH']/parent::div/following-sibling::p").text
    except NoSuchElementException:
        jh_review = ''

    try:
        v_review = driver.find_element_by_xpath("//div[@class='pipProfessionalReviews_rating']/span[@class='wineRatings_initials'][text() = 'V']/parent::div/following-sibling::p").text
    except NoSuchElementException:
        v_review = ''

    try:
        wns_review = driver.find_element_by_xpath("//div[@class='pipProfessionalReviews_rating']/span[@class='wineRatings_initials']/span[@class = 'wineRatings_initialsAmpersand']/parent::span/parent::div/following-sibling::p").text
    except NoSuchElementException:
        wns_review = ''

    # Ratings
    try:
        ww_rating = driver.find_element_by_xpath("//div[@class='prodInfo']/ul[@class='wineRatings_list']/li/span[@class='wineRatings_initials'][text() = 'WW']/parent::div/following-sibling::p").text
    except NoSuchElementException:
        ww_rating = ''

    try:
        ws_rating = driver.find_element_by_xpath("//div[@class='prodInfo']/ul[@class='wineRatings_list']/li/span[@class='wineRatings_initials'][text() = 'WS']/parent::div/following-sibling::p").text
    except NoSuchElementException:
        ws_rating = ''

    try:
        we_rating = driver.find_element_by_xpath("//div[@class='prodInfo']/ul[@class='wineRatings_list']/li/span[@class='wineRatings_initials'][text() = 'WE']/parent::div/following-sibling::p").text
    except NoSuchElementException:
        we_rating = ''

    try:
        tp_rating = driver.find_element_by_xpath("//div[@class='prodInfo']/ul[@class='wineRatings_list']/li/span[@class='wineRatings_initials'][text() = 'TP']/parent::div/following-sibling::p").text
    except NoSuchElementException:
        tp_rating = ''

    try:
        cg_rating = driver.find_element_by_xpath("//div[@class='prodInfo']/ul[@class='wineRatings_list']/li/span[@class='wineRatings_initials'][text() = 'CG']/parent::div/following-sibling::p").text
    except NoSuchElementException:
        cg_rating = ''

    try:
        rp_rating = driver.find_element_by_xpath("//div[@class='prodInfo']/ul[@class='wineRatings_list']/li/span[@class='wineRatings_initials'][text() = 'RP']/parent::div/following-sibling::p").text
    except NoSuchElementException:
        rp_rating = ''

    try:
        wr_rating = driver.find_element_by_xpath("//div[@class='prodInfo']/ul[@class='wineRatings_list']/li/span[@class='wineRatings_initials'][text() = 'WP']/parent::div/following-sibling::p").text
    except NoSuchElementException:
        wr_rating = ''

    try:
        d_rating = driver.find_element_by_xpath("//div[@class='prodInfo']/ul[@class='wineRatings_list']/li/span[@class='wineRatings_initials'][text() = 'D']/parent::div/following-sibling::p").text
    except NoSuchElementException:
        d_rating = ''

    try:
        js_rating = driver.find_element_by_xpath("//div[@class='prodInfo']/ul[@class='wineRatings_list']/li/span[@class='wineRatings_initials'][text() = 'JS']/parent::div/following-sibling::p").text
    except NoSuchElementException:
        js_rating = ''

    try:
        jh_rating = driver.find_element_by_xpath("//div[@class='prodInfo']/ul[@class='wineRatings_list']/li/span[@class='wineRatings_initials'][text() = 'JH']/parent::div/following-sibling::p").text
    except NoSuchElementException:
        jh_rating = ''

    try:
        v_rating = driver.find_element_by_xpath("//div[@class='prodInfo']/ul[@class='wineRatings_list']/li/span[@class='wineRatings_initials'][text() = 'V']/parent::div/following-sibling::p").text
    except NoSuchElementException:
        v_rating = ''

    try:
        wns_rating = driver.find_element_by_xpath("//div[@class='prodInfo']/ul[@class='wineRatings_list']/li/span[@class='wineRatings_initials']/span[@class = 'wineRatings_initialsAmpersand']/parent::span/parent::div/following-sibling::p").text
    except NoSuchElementException:
        wns_rating = ''

    data = [str(count), str(i), wine_name, wine_varietal, wine_origin, winery_name, winery_description, alcohol_level, original_price, discounted_price,
            saving, sold_out, est_price, collectible, red_wine, green_wine, white_wine, great_gift, region_name, region_info, longitude, latitude,
            rating, no_of_reviews, winemaker_notes, ws_review, wr_review, rp_review, cg_review, wns_review, ww_review,
            we_review, tp_review, js_review, d_review, jh_review, v_review, ws_rating, wr_rating, rp_rating, cg_rating,
            wns_rating, ww_rating, we_rating, tp_rating, js_rating, d_rating, jh_rating, v_rating]

    # print(data)
    header_first_vintage = headers.index("2016_PROD_ID")
    for k in range(1, len(headers)):
        data.append("")

    try:
        driver.find_element_by_css_selector(".prodPedigree_allLink").click()
        vintages = driver.find_elements_by_xpath(
            "//div[@class='pedigreeList_container']/div[contains(@class,'prodPedigree_flexContainer')]")
    except NoSuchElementException:
        vintages = driver.find_elements_by_xpath(
            "//div[@class='pedigreeList_container']/div[contains(@class,'prodPedigree_flexContainer')]")
    except ElementNotVisibleException:
        vintages = []
    vintages_data = []
    if len(vintages) == 0:
        try:
            vintages = driver.find_elements_by_xpath("//div[contains(@class,'prodPedigree_flexContainer')]")
        except (NoSuchElementException, ElementNotVisibleException):
            vintages = []
    for index, vintage in enumerate(vintages):
        year = vintage.find_element_by_xpath("./label").text
        header_first_vintage = headers.index(year + "_PROD_ID")
        try:
            prod_id = vintage.find_element_by_xpath("./label/a").get_attribute('href').split("/")[-1]
        except AttributeError:
            prod_id = ''
        try:
            ws_rating = vintage.find_element_by_xpath("./div/ul/li/span[text() = 'WS']/following-sibling::span").text
        except NoSuchElementException:
            ws_rating = ''

        try:
            ww_rating = vintage.find_element_by_xpath("./div/ul/li/span[text() = 'WW']/following-sibling::span").text
        except NoSuchElementException:
            ww_rating = ''

        try:
            we_rating = vintage.find_element_by_xpath("./div/ul/li/span[text() = 'WE']/following-sibling::span").text
        except NoSuchElementException:
            we_rating = ''

        try:
            rp_rating = vintage.find_element_by_xpath("./div/ul/li/span[text() = 'RP']/following-sibling::span").text
        except NoSuchElementException:
            rp_rating = ''

        try:
            cg_rating = vintage.find_element_by_xpath("./div/ul/li/span[text() = 'CG']/following-sibling::span").text
        except NoSuchElementException:
            cg_rating = ''

        try:
            js_rating = vintage.find_element_by_xpath("./div/ul/li/span[text() = 'JS']/following-sibling::span").text
        except NoSuchElementException:
            js_rating = ''

        try:
            wr_rating = vintage.find_element_by_xpath("./div/ul/li/span[text() = 'WR']/following-sibling::span").text
        except NoSuchElementException:
            wr_rating = ''

        try:
            wns_rating = vintage.find_element_by_xpath("./div//ul/li/span/span[@class = 'wineRatings_initialsAmpersand']"
                                                       "/parent::span/following-sibling::span").text
        except NoSuchElementException:
            wns_rating = ''

        try:
            tp_rating = vintage.find_element_by_xpath("./div/ul/li/span[text() = 'TP']/following-sibling::span").text
        except NoSuchElementException:
            tp_rating = ''

        try:
            d_rating = vintage.find_element_by_xpath("./div/ul/li/span[text() = 'D']/following-sibling::span").text
        except NoSuchElementException:
            d_rating = ''

        try:
            jh_rating = vintage.find_element_by_xpath("./div/ul/li/span[text() = 'JH']/following-sibling::span").text
        except NoSuchElementException:
            jh_rating = ''

        try:
            v_rating = vintage.find_element_by_xpath("./div/ul/li/span[text() = 'V']/following-sibling::span").text
        except NoSuchElementException:
            v_rating = ''
        vintage_data = [prod_id, ws_rating, wr_rating, rp_rating, cg_rating, wns_rating, ww_rating, we_rating, tp_rating, js_rating,
                        d_rating, jh_rating, v_rating]
        # vintages_data.extend(vintage_data)
        data[header_first_vintage:header_first_vintage + 14] = vintage_data

    with open('data.csv', 'a', newline='', encoding='utf-8') as f:
        writer = csv.writer(f)
        writer.writerow(data)

    time.sleep(random.randint(5, 10))
```
